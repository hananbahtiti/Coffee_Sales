```python
# Define the class for data preprocessing tasks
class DataPreprocessing():

    # Constructor method to initialize class
    def __init__(self) -> None:
        self.__df = None  # Private attribute to store the dataset

    # Private method to download/load the dataset
    def __download_dataset(self):
        if self.__df is not None:  # If dataset already loaded, return it
            return self.__df

        try:
            file_path = "Coffee Shop Sales.xlsx"  # File name of dataset

            # Load the dataset using kagglehub
            self.__df = kagglehub.load_dataset(
                KaggleDatasetAdapter.PANDAS,
                "ahmedabbas757/coffee-sales",  # Dataset source
                file_path
            )
            return self.__df  # Return the loaded dataset

        except Exception as e:
            return f"An error occurred: {e}"  # Return error message if exception occurs

    # Method to read dataset info and top rows
    def read_dataset(self, rows_number=5):
        try:
            df = self.__download_dataset()  # Load dataset
            info = df.info()  # Get summary info of dataset (printed to console)
            head = df.head()  # Get first 5 rows
            return info, head  # Return info and head

        except Exception as e:
            return f"An error occurred: {e}"  # Return error message

    # Method to show object-type columns and their unique values
    def show_data(self):
        try:
            df = self.__download_dataset()  # Load dataset
            cols = df.columns.tolist()  # Get list of column names
            object_columns = [col for col in cols if df[col].dtype == 'object']  # Filter object-type columns
            unique_value = df[object_columns].nunique()  # Count unique values for each object column
            return unique_value  # Return the counts

        except Exception as e:
            return f"An error occurred: {e}"  # Return error message

    # Method to display value counts for all columns
    def value_count(self):
        try:
            df = self.__download_dataset()  # Load dataset
            columns_names = df.columns  # Get all column names
            for column in columns_names:
                print(f"========================= {column} =========================")
                print(f"{df[column].value_counts()}")  # Print value counts for each column

        except Exception as e:
            return f"An error occurred: {e}"  # Return error message

# Create an instance of the DataPreprocessing class
coffee_sales = DataPreprocessing()
```

```python
file_path = "Coffee Shop Sales.xlsx"

      # Load the dataset
df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "ahmedabbas757/coffee-sales",
    file_path
)
```

```python
# The first line identifies missing (NaN) values across all columns to assess data completeness.

df.isna().sum()
```

```python
# Group the dataset by product category and product type,
# count the number of occurrences for each combination,
# and sort the results in descending order to identify the most frequently sold products.
df.groupby(['product_category', 'product_type']).size().sort_values(ascending=False)
```

```python
# This line aggregates the number of transactions by transaction date and product category,
# enabling the identification of peak sales periods per category.
# The results are sorted in descending order to highlight the highest sales activity first.
df.groupby(['transaction_date', 'product_category']).size().sort_values(ascending=False)
```

```python
# Convert the transaction_date column to monthly periods in string format,
# then group the data by month and product category to calculate the number of transactions per combination.
# The results are sorted in descending order to identify the most active product categories by month.
df_transaction_date = df['transaction_date'].dt.to_period('M').astype('str').sort_values(ascending=False)
df.groupby([df_transaction_date, 'product_category']).size().sort_values(ascending=False)
```

```python
# This section converts the transaction_date column into monthly periods as strings,
# then groups the data by transaction month, product category, and product type.
# It calculates the frequency of each product type sold per category and month,
# and sorts the results in descending order to highlight the top-selling items over time.

df_transaction_date = df['transaction_date'].dt.to_period('M').astype('str').sort_values(ascending=False)
df.groupby([df_transaction_date, 'product_category', 'product_type']).size().sort_values(ascending=False)
```

```python
# This line groups the dataset by the quantity of items in each transaction and the store location.
# It calculates the frequency of each quantity-location combination to identify which store locations
# commonly process transactions of certain quantities.
# The results are sorted in descending order to highlight the most frequent transaction volumes by location.

df.groupby(['transaction_qty' ,'store_location']).size().sort_values(ascending=False)
```

```python
# This section extracts the hour from each transaction time and groups the data by hour, product category, and product type.
# It calculates the number of transactions for each combination to identify which product types are most frequently sold at specific hours of the day.
# The results are sorted in descending order to highlight peak sales hours and top-performing products during those times.

df_transaction_time = pd.to_datetime(df['transaction_time'], format='%H:%M:%S').dt.hour.sort_values(ascending=False)
df.groupby([df_transaction_time, 'product_category', 'product_type']).size().sort_values(ascending=False)
```

```python
# This analysis groups the data by product category, product type, and unit price to determine how frequently each product-price combination appears.
# It provides insights into the most commonly sold items and their price points, helping identify top-performing products and the most attractive pricing strategies.
# The results are sorted in descending order to highlight the highest-selling combinations for strategic planning and pricing optimization.

df.groupby(['product_category', 'product_type' ,'unit_price']).size().sort_values(ascending=False)
```

```python
# This grouping analyzes sales performance at the most granular level by considering product category, type, and specific product detail.
# It identifies the most frequently sold product variants, offering insight into customer preferences at a detailed level.
# Results are sorted in descending order to prioritize top-selling product combinations, supporting product planning, inventory management, and marketing focus.

df.groupby(['product_category', 'product_type' ,'product_detail']).size().sort_values(ascending=False)
```
